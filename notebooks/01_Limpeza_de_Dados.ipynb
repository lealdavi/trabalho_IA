```python
tickers_b3 = [
    # Blue chips e grandes bancos
    'PETR4.SA', 'VALE3.SA', 'ITUB4.SA', 'BBDC4.SA', 'ABEV3.SA',
    'RENT3.SA', 'WEGE3.SA', 'BBAS3.SA', 'SUZB3.SA', 'BRFS3.SA',

    # Mineração e Siderurgia
    'CSNA3.SA', 'GGBR4.SA', 'CMIG4.SA', 'EGIE3.SA', 'RAIL3.SA',

    # Varejo e Consumo
    'LREN3.SA', 'MGLU3.SA', 'CVCB3.SA', 'PCAR3.SA', 'AZUL4.SA',
    'HAPV3.SA', 'VIVT3.SA', 'B3SA3.SA', 'FLRY3.SA', 'TOTS3.SA',
    'RADL3.SA', 'RDOR3.SA', 'JBSS3.SA', 'MRFG3.SA', 'ASAI3.SA',

    # Energia e Saneamento
    'AMBP3.SA', 'ARZZ3.SA', 'BRSR6.SA', 'CCRO3.SA', 'CIEL3.SA',
    'CPFE3.SA', 'CPLE6.SA', 'CRFB3.SA', 'CSMG3.SA', 'CYRE3.SA',

    # Imobiliário e Construção
    'DXCO3.SA', 'ELET3.SA', 'EMBR3.SA', 'ENBR3.SA', 'ENGI4.SA',
    'EQTL3.SA', 'ENAT3.SA', 'GOAU4.SA', 'GOLL4.SA',

    # Saúde e Tecnologia
    'HYPE3.SA', 'IGTI11.SA', 'IRBR3.SA', 'ITSA4.SA', 'KLBN4.SA',
    'LWSA3.SA', 'MOVI3.SA', 'MRVE3.SA', 'MULT3.SA',

    # Diversos (Petróleo, Químicos, Transporte)
    'NATU3.SA', 'PETZ3.SA', 'PRIO3.SA', 'QUAL3.SA', 'RRRP3.SA',
    'SANB4.SA', 'SBSP3.SA', 'TAEE4.SA', 'TIMS3.SA', 'UGPA3.SA',
    'USIM5.SA', 'VBBR3.SA', 'VIIA3.SA', 'YDUQ3.SA',  # <-- VVAR3 foi substituído por VIIA3

    # Adicionando mais alguns robustos
    'ALPA4.SA', 'BAZA3.SA', 'BEEF3.SA', 'BPAN4.SA', 'BRAP4.SA',
    'BRKM5.SA', 'CMIN3.SA', 'COGN3.SA', 'CSAN3.SA', 'FRAS3.SA',
    'GRND3.SA', 'HBOR3.SA', 'JHSF3.SA'
]

# ordena tickers
tickers_b3 = sorted(set(tickers_b3))
print(f" Total de tickers definidos: {len(tickers_b3)}")
print(f" Primeiros 10 tickers: {tickers_b3[:10]}")
```

```python
data_fim = datetime.now()
data_inicio = data_fim - timedelta(days=15*365)

print(f"\n Período de coleta: {data_inicio.strftime('%d/%m/%Y')} até {data_fim.strftime('%d/%m/%Y')}")

dados_brutos = yf.download(
    tickers=tickers_b3,
    start=data_inicio,
    end=data_fim,
    group_by='ticker', # agrupa dados por cada ativo
    auto_adjust=True, # corrije splits e dividendos
    progress=True
)
```

```python
# Verifica se o download retornou dados
if dados_brutos.empty:
    raise ValueError("ERRO: Nenhum dado foi baixado. Verifique sua conexão ou a lista de tickers.")

# Extrai a coluna 'Close' lidando com MultiIndex ou DataFrame simples
if isinstance(dados_brutos.columns, pd.MultiIndex):
    # MultiIndex: (Ticker, Atributo) -> pegamos o nível 'Close'
    precos_acoes = dados_brutos.xs('Close', axis=1, level=1)
else:
    # Caso único ticker ou formato diferente
    if 'Close' in dados_brutos.columns:
        precos_acoes = dados_brutos[['Close']] if isinstance(dados_brutos, pd.DataFrame) else pd.DataFrame(dados_brutos['Close'])
    else:
        # Fallback: assume que o DataFrame já é só preços
        precos_acoes = dados_brutos

# Garantir que é um DataFrame (se for Series, converte)
if isinstance(precos_acoes, pd.Series):
    precos_acoes = pd.DataFrame(precos_acoes)
```

```python
print("\n" + "-"*60)
print("RELATORIO DE DOWNLOAD")
print("-"*60)

tickers_baixados = precos_acoes.columns.tolist()
total_baixados = len(tickers_baixados)
total_lista = len(tickers_b3)

print(f">> Tickers na lista original: {total_lista}")
print(f">> Tickers baixados com sucesso: {total_baixados}")

# Identificar quais falharam
set_baixados = set(tickers_baixados)
set_lista = set(tickers_b3)
falharam = set_lista - set_baixados

if falharam:
    print(f">> [ATENCAO] Tickers que falharam (serao ignorados): {len(falharam)}")
    print(f"   Lista: {sorted(list(falharam))}")
else:
    print(">> [OK] Todos os tickers da lista foram baixados com sucesso!")
```

```python
print("\n" + "-"*60)
print("RELATORIO DE LIMPEZA")
print("-"*60)

# Verifica percentual de dados faltantes por ticker
dias_totais = len(precos_acoes)
limite_minimo = 0.70  # Pelo menos 70% dos dias precisam ter preco
min_dias_necessarios = int(limite_minimo * dias_totais)

print(f">> Dias totais no periodo: {dias_totais}")
print(f">> Dias minimos exigidos (70%): {min_dias_necessarios}")

# Conta quantos dados faltantes cada ação tem antes da limpeza
missing_before = precos_acoes.isnull().sum()
tickers_com_missing = missing_before[missing_before > 0]

if not tickers_com_missing.empty:
    print(f">> Acoes com dados faltantes antes da limpeza: {len(tickers_com_missing)}")
    print(f"   Media de dias faltantes: {tickers_com_missing.mean():.1f}")
else:
    print(">> Nenhuma acao com dados faltantes antes da limpeza.")

# Remove acoes com mais de 30% de dados faltantes
precos_acoes_limpos = precos_acoes.dropna(thresh=min_dias_necessarios, axis=1)

# Identifica quais foram removidos
removidos_limpeza = set(precos_acoes.columns) - set(precos_acoes_limpos.columns)
if removidos_limpeza:
    print(f">> [ATENCAO] Acoes removidas por baixa liquidez (>30% faltantes): {len(removidos_limpeza)}")
    print(f"   Lista: {sorted(list(removidos_limpeza))}")
else:
    print(">> [OK] Nenhuma acao removida por baixa liquidez.")

# Preenche os poucos dados faltantes restantes (forward fill + backward fill)
precos_acoes_limpos = precos_acoes_limpos.ffill().bfill()

# Verifica se ainda restam dados faltantes (alerta)
missing_after = precos_acoes_limpos.isnull().sum().sum()
if missing_after > 0:
    print(f">> [ATENCAO] Ainda restam {missing_after} dados faltantes apos o preenchimento.")
else:
    print(">> [OK] Nenhum dado faltante remanescente apos o preenchimento.")

print(f"\n>> Total de acoes na base final: {precos_acoes_limpos.shape[1]}")
```

```python
print("\n" + "-"*60)
print("ANALISE EXPLORATORIA DOS DADOS LIMPOS")
print("-"*60)

print(f">> Shape: {precos_acoes_limpos.shape[0]} dias x {precos_acoes_limpos.shape[1]} acoes")
print(f">> Periodo: {precos_acoes_limpos.index[0].strftime('%d/%m/%Y')} ate {precos_acoes_limpos.index[-1].strftime('%d/%m/%Y')}")

# Estatistica descritiva basica (precos)
print("\n>> Estatistica descritiva dos precos (todas as acoes):")
display(precos_acoes_limpos.describe().T.round(2).head(10))

# Verificacao de tipos e memoria
print("\n>> Informacoes gerais do DataFrame:")
print(f"   Tipos das colunas: {precos_acoes_limpos.dtypes.value_counts().to_dict()}")
print(f"   Memoria utilizada: {precos_acoes_limpos.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```

```python
caminho_dataset = f'../data/processed/'

# Cria a pasta se nao existir
if not os.path.exists(caminho_dataset):
    os.makedirs(caminho_dataset)
    print(f">> Pasta criada: {caminho_dataset}")
else:
    print(f">> Pasta ja existe: {caminho_dataset}")

# Salvar a base de precos limpa
arquivo_precos = os.path.join(caminho_dataset, 'stocks_b3.csv')
precos_acoes_limpos.to_csv(arquivo_precos)
print(f">> [OK] Base de precos salva: {arquivo_precos}")

# Salvar a lista de tickers baixados (sucesso)
arquivo_sucessos = os.path.join(caminho_dataset, 'tickers_baixados.csv')
pd.Series(sorted(tickers_baixados)).to_csv(arquivo_sucessos, index=False, header=['ticker'])
print(f">> [OK] Lista de tickers baixados salva: {arquivo_sucessos}")

# Salvar a lista de tickers que falharam no download (para documentacao)
if falharam:
    arquivo_falhas = os.path.join(caminho_dataset, 'tickers_falharam.csv')
    pd.Series(sorted(list(falharam))).to_csv(arquivo_falhas, index=False, header=['ticker'])
    print(f">> [OK] Lista de tickers que falharam salva: {arquivo_falhas}")

# Salvar a lista de tickers removidos na limpeza (para documentacao)
if removidos_limpeza:
    arquivo_removidos = os.path.join(caminho_dataset, 'tickers_removidos_limpeza.csv')
    pd.Series(sorted(list(removidos_limpeza))).to_csv(arquivo_removidos, index=False, header=['ticker'])
    print(f">> [OK] Lista de tickers removidos na limpeza salva: {arquivo_removidos}")

# Salvar um relatorio resumo em TXT
relatorio = f"""
RELATORIO DE INGESTAO E LIMPEZA
===============================
Data de execucao: {datetime.now().strftime('%d/%m/%Y %H:%M:%S')}
Periodo dos dados: {precos_acoes_limpos.index[0].strftime('%d/%m/%Y')} a {precos_acoes_limpos.index[-1].strftime('%d/%m/%Y')}
Total de dias: {precos_acoes_limpos.shape[0]}
----------------------------------------
Tickers na lista original: {total_lista}
Tickers baixados com sucesso: {total_baixados}
Tickers que falharam no download: {len(falharam)}
Tickers removidos por baixa liquidez: {len(removidos_limpeza)}
Tickers na base final: {precos_acoes_limpos.shape[1]}
----------------------------------------
Arquivos gerados:
- precos_acoes_b3_limpos.csv (dados principais)
- tickers_baixados.csv
- tickers_falharam.csv (se houver)
- tickers_removidos_limpeza.csv (se houver)
"""
arquivo_relatorio = os.path.join(caminho_dataset, 'relatorio_ingestao.txt')
with open(arquivo_relatorio, 'w', encoding='utf-8') as f:
    f.write(relatorio)
print(f">> [OK] Relatorio resumo salvo: {arquivo_relatorio}")
```
